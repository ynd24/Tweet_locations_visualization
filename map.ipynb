```python
import pandas as pd
```

```python
df = pd.read_excel('Merged_Final.xlsx') 
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
df.head(10)
```

```python
df1 = df[['Gender', 'user_location']] # new df with only relavent columns
```

```python
df1.head(5)
```

```python
df1['Gender'] = df1['Gender'].astype('string') # change dtype to string
df1['user_location'] = df1['user_location'].astype('string') # change dtype to string
```

```python
df1['Gender'] = df1['Gender'].str.lower() # lower case all genders
df1['Gender'] = df1['Gender'].str.strip() # remove trailing and leading whitespace
df1['user_location'] = df1['user_location'].str.strip() 
```

```python
df1['user_location'] = df1['user_location'].replace({'USA': 'United States of America', 'United States': 'United States of America'})
# Countries in my DF must match countries in countries.geojson
```

```python
grouped_data = df1.groupby(['Gender', 'user_location']).size().reset_index(name='Count') # group data by two coloumns
```

```python
pivot_data = grouped_data.pivot(index='user_location', columns='Gender', values='Count').reset_index()
```

```python
pivot_data.columns.name = None  # Remove the column name
pivot_data.rename(columns={'female': 'Female', 'male': 'Male', 'none': 'None'}, inplace=True) # add new coloumn names
pivot_data.fillna(0, inplace=True) # fill all null vals with 0
```

```python
pivot_data['Total'] = pivot_data['Male']+pivot_data['Female']+pivot_data['None'] # create new col for total tweets
```

```python
pivot_data
```

```python
pivot_data= pivot_data[pivot_data['user_location'] != 'None'] # remove tweets which had no country,19920.0 tweets
```

```python
pivot_data.to_csv('Countries_num_tweets') 
```

```python
import folium
import pandas as pd


# Create a Folium map
m = folium.Map(location=[0, 0], zoom_start=2)

# Define custom color bins and their ranges
colors = ['green', 'yellow', 'red']
threshold_scale = [0,50,150,300,450,600,750,900,pivot_data['Female'].max()]  # range of scale

# Add choropleth layer for Female with custom color scale and ranges
folium.Choropleth(
    geo_data="countries.geojson",  # GeoJSON file with country boundaries
    name="Female",
    data=pivot_data,
    columns=["user_location", "Female"],
    key_on="feature.properties.ADMIN",
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name="# of Female Tweets",
    bins=threshold_scale,  # Use the custom threshold scale
    threshold_scale=threshold_scale,  # Specify the custom threshold scale
    fill_color='YlGnBu',  # Use a base color scale without a lambda function
).add_to(m)


# Display the map
m.save("female_custom_colors.html")
```
